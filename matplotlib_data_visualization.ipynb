```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'sans-serif' 
```

```python
import seaborn as sns
sns.get_dataset_names()
```

```python
dowjones = sns.load_dataset("dowjones")
dowjones.head()
```

```python
dowjones.dtypes
```

```python
dowjones.groupby("Date").mean()
```

#### Line Plot

```python
plt.figure(figsize=(12,6), dpi = 300)
plt.plot(dowjones["Date"], dowjones["Price"], color= "red", linestyle= "--", marker= "o", label= "Line")
plt.title("Line Plot", fontsize= 14, fontweight= "bold", color= "black", loc= "center")
plt.xlabel("Date", fontsize= 12)
plt.ylabel("Price", fontsize= 12)
plt.grid(True, linestyle= "--", linewidth = 0.5, alpha= 0.5)
plt.show()
```

#### Bar chart

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
plt.figure(figsize=(5,3), dpi=100)
sex_proportion = tips["sex"].value_counts()
plt.bar(sex_proportion.index,sex_proportion.values, color= "skyblue", width=0.8)
plt.title("Sex proportion of Waiters", fontsize=14, fontweight="bold", color="black",loc="center")
plt.xlabel("Sex", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.grid(axis="y", color="gray", ls="--", lw=0.5, alpha=0.5)
plt.xticks(rotation=45)
plt.show()
```

```python
taxis = sns.load_dataset("taxis")
taxis.head()
```

```python
pickup_zone_count = taxis["pickup_zone"].value_counts().nlargest(10)
plt.figure(figsize=(12,6))
plt.barh(pickup_zone_count.index, pickup_zone_count.values, color="green", height=0.8, edgecolor="black",alpha=0.5 )
plt.title("Top 10 Pickup_borough", fontsize=14, fontweight="bold", color="black", loc="center")
plt.xlabel("Count", fontsize=12)
plt.ylabel("Pickup_Borough", fontsize=12)
plt.grid(axis="x", ls="--", lw=0.5, alpha=0.5)
plt.show()
```

#### 

```python
day_sex_count = pd.crosstab(tips["day"], tips["sex"])
plt.bar(day_sex_count.index, day_sex_count.values[:,0], color="darkblue", label="Male")
plt.bar(day_sex_count.index, day_sex_count.values[:,1],bottom= day_sex_count.values[:,0], color="green", label="Female")
plt.title("Day by Gender", fontsize=14)
plt.xlabel("Day")
plt.ylabel("Count")
plt.legend(title="Gender")
plt.show()
```

### Scatter Plot

```python
plt.scatter(tips["total_bill"], tips["tip"], color="green", s=50, alpha=0.8, ec="black", lw=0.1)
```

### Histogram

```python
plt.hist(tips["tip"], bins=20, color="skyblue", edgecolor="black", histtype="bar")
plt.show()
```

```python
plt.hist(tips["total_bill"], alpha = 0.5, label = 'Total Bill')
plt.hist(tips["tip"], alpha = 0.5, label = 'Tip')
plt.legend()
plt.show()
```

```python
### Histogram in Subplot
fig, ax = plt.subplots(1,2, figsize = (10,5))
ax[0].hist(tips["tip"], color="skyblue")
ax[0].set_title('Histogram')
ax[1].hist(tips["tip"], color="skyblue", cumulative = True)
ax[1].set_title('Cumulative Histogram')
plt.tight_layout()
plt.show()
```

### Pie chart

```python
plt.pie(tips["day"].value_counts(), labels=tips["day"].value_counts().index, autopct="%1.2f%%",
        colors = ['yellow', 'hotpink', 'b', '#4CAF50'], startangle=90, explode=[0.05,0,0,0], shadow=False)
plt.show()
```

```python
day_sex_count = pd.crosstab(tips["day"], tips["sex"])
day_sex_count
```

```python
day_sex_count = pd.crosstab(tips["day"], tips["sex"])
fig, axes = plt.subplots(1, 2, figsize=(10, 5))  # 1 row, 2 columns

# First pie chart
axes[0].pie(day_sex_count.values[:,0], labels=day_sex_count.index, autopct="%1.2f%%",
        colors = ['yellow', 'hotpink', 'b', '#4CAF50'], startangle=90, explode=[0,0,0,0.05], shadow=False)
axes[0].set_title("Male")

# Second pie chart
axes[1].pie(day_sex_count.values[:,1], labels=day_sex_count.index, autopct="%1.2f%%",
        colors = ['yellow', 'hotpink', 'b', '#4CAF50'], startangle=90, explode=[0,0,0,0.05], shadow=False)
axes[1].set_title("Female")

plt.tight_layout()
plt.show()
```

#### Boxplot

```python
plt.figure(figsize=(12,6))
plt.boxplot(tips["tip"])
plt.title("Distribution - Outliers")
plt.xticks([1], ["tip"])
plt.grid(linestyle="--", alpha=0.5)
plt.show()
```
