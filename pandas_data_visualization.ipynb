```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
sns.get_dataset_names()
```

```python
dowjones= sns.load_dataset("dowjones")
dowjones.head()
```

### Line Graph

```python
dowjones.plot(x= "Date", y= 'Price', kind= 'line',figsize=(12,6), marker= 'o', color= 'darkblue',
       title= 'Daily Price Trend', xlabel= 'Date', ylabel = 'Price',legend= False)
plt.show()
```

```python
dowjones.plot(x= "Date", y= 'Price', kind= 'area',figsize=(12,6),
       title= 'Daily Price Trend', xlabel= 'Date', ylabel = 'Price', legend=False)
plt.show()
```

### Bar Graph

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sex_count= tips["sex"].value_counts()
sex_count
```

```python
sex_count.plot(x = sex_count.index, y = sex_count.values, kind = 'bar', color= 'skyblue',
               title = 'Gender Distribution', xlabel = 'Gender', ylabel = 'Count', legend = False, rot=0)
plt.show()
```

```python
sex_count.sort_values(ascending = True).\
plot(x = sex_count.index, y = sex_count.values, kind = 'barh', title = "Gender Distribution", 
     xlabel = 'Count', ylabel = 'Gender', color = 'skyblue', legend = False)
plt.show()
```

```python
sex_smoker = pd.crosstab(tips["sex"], tips["smoker"]).reset_index()
sex_smoker
```

```python
sex_smoker.plot(x= "sex", kind= 'bar', title= 'Smoker by Gender',color=["#80a7c6", "#9478c5"],
       xlabel= "Gender", ylabel= 'Count', rot= 0)
plt.show()
```

```python
sex_smoker.plot(x= "sex", kind= 'barh', title= 'Smoker by Gender',color=["#80a7c6", "#9478c5"],
       xlabel= "Count", ylabel= 'Gender', rot= 0)
plt.show()
```

```python
sex_smoker.plot(x= "sex", kind= 'bar', stacked=True, title= 'Smoker by Gender',color=["#80a7c6", "#9478c5"],
       xlabel= "Gender", ylabel= 'Count', rot= 0)
plt.show()
```

```python
sex_smoker.plot(x= "sex", kind= 'barh', stacked=True, title= 'Smoker by Gender',color=["#80a7c6", "#9478c5"],
       xlabel= "Count", ylabel= 'Gender', rot= 0)
plt.show()
```

### Histogram

```python
tips["tip"].plot(kind = 'hist', title = 'Tip Distribution', color = 'skyblue',
       bins = 15, edgecolor = 'black')
plt.show()
```

### Box Plot

```python
tips["tip"].plot(kind= 'box', title= 'Tip Distribution and Outliers')
plt.show()
```

### KDE (Kernal Density Estimation) Plot

```python
tips["tip"].plot(kind= 'kde', color= 'green')
plt.show()
```

### Scatter Plot

```python
tips.plot(x= "total_bill", y= "tip", kind= 'scatter', color="g", figsize=(12,6), 
          title= 'Relationship between total bill and tip')
plt.show()
```

```python
tips.plot(kind="hexbin", x="total_bill", y="tip", gridsize=20)
```

#### Pie chart

```python
sex_count = tips["sex"].value_counts()
sex_count
```

```python
sex_count.plot(y=sex_count.values, kind="pie", labels=sex_count.index, startangle=90, figsize=(6,4), 
               colors=["skyblue", "lightgreen"], autopct="%.1f%%", explode=[0.1,0], title="Gender Proportion", 
               ylabel="", legend=False)
plt.show()
```
