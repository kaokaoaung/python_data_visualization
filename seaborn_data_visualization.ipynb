```python
sns.pairplot(tips, hue = "sex", diag_kind = "kde")
```

#### Subplots

```python
iris=sns.load_dataset("iris")
iris.head()
```

```python
fig, ax= plt.subplots(2,2, figsize=(10,6))
sns.violinplot(data= iris, x= 'species', y= 'sepal_length', ax= ax[0,0])
sns.violinplot(data= iris, x= 'species', y= 'sepal_width', ax= ax[0,1])
sns.violinplot(data= iris, x= 'species', y= 'petal_length', ax= ax[1,0])
sns.violinplot(data= iris, x= 'species', y= 'petal_width', ax= ax[1,1])
plt.show()
```

```python
fig, ax= plt.subplots(2,2, figsize= (10,6), constrained_layout= True)
iris.sepal_length.plot.kde(ax= ax[0,0], color= 'green')
ax[0,0].set_title("Sepal Length")
iris.sepal_width.plot.kde(ax= ax[0,1], color= 'blue')
ax[0,1].set_title("Sepal Width")
iris.petal_length.plot.kde(ax= ax[1,0], color= 'red')
ax[1,0].set_title('Petal Length')
iris.petal_width.plot.kde(ax= ax[1,1], color= 'magenta')
ax[1,1].set_title('Petal Width')
plt.show()
```
